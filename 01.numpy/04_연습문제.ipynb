#### 5. 0에서 10까지 랜덤 실수값으로 이루어진 5x6 형태의 데이터 행렬

#### 6. 특정 행 기준 재정렬

```python
b = np. array([[ 1, 2, 3, 4],
[ 46, 99, 100, 71],
[ 81, 59, 90, 100]])

b[:, b[1, :].argsort()]
```

```python
# 오름차순 한꺼번에 수행
b[:, np.argsort(b[1])]
```

```python
# 내림차순 한꺼번에 수행
b[:, np.argsort(b[1])[::-1]]
```

#### 8. 10,000원인 주식, 일간 수익률은 기댓값은 0%, 표준편차가 1%인 표준 정규 분포를 따른다. 250일 동안의 주가 무작위 생성.

```python
r = np.random.normal(0, 1, 250)

price = 10000
for i in range(250):
    price += r
print(price)
```

```python
# 풀이
np.random.seed(2021)
price = 10000
rate = np.random.normal(0, 1, 10)
for i in range(10):
    price += np.round(price * rate[i] * 0.01)
    print(price, end=', ')
```

```python
np.random.seed(2021)
price_list = []
rate = np.random.normal(0, 1, 250)
price = 10000
for i in range(250):
    price += np.round(price * rate[i] * 0.01)
    price_list.append(price)
price_array = np.array(price_list, int)
```

```python
print(price_array[:10])
print(price_array[-10:])
```
