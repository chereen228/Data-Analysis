### 여러 개의 그림 그리기

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams["font.family"] = 'Malgun Gothic'
```

#### 파이썬 랜덤모듈
- np.random.randint

```python
np.random.randint(6)        # 0부터 5까지 랜덤한 숫자 1개
```

```python
np.random.seed(2021)            # 고정
np.random.randint(1, 20)        # 1부터 19까지 랜덤한 숫자 1개
```

- np.random.rand(m, n)

```python
np.random.rand(6)       # 0~1의 표준정규분포 난수 array 생성 
```

```python
np.random.rand(3, 2)        # 3x2 배열로 생성
```

```python
np.random.randn(6)      # 평균0, 표준편차1의 가우시안 표준정규분포 난수 6개 생성 -> array
```

```python
np.random.randn(3, 2)       # 3x2 배열로 생성
```

- Figure 객체

```python
# figsize: 그림 크기 설정
np.random.seed(2021)
plt.figure(figsize=(10, 2))
plt.title('figure size: (10, 2)')
plt.plot(np.random.randn(100))
plt.grid()
plt.show()
```

- Axes 객체와 subplot

```python
x1 = np.linspace(0., 5.)    # num 지정하지 않으면 '50'이 디폴트 값
x2 = np.linspace(0., 2.)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)
```

```python
ax1 = plt.subplot(2, 1, 1)      # 2 x 1 행렬 위쪽 플롯
plt.plot(x1, y1, 'yo-')
plt.title('2 Subplot')
plt.ylabel('Damped oscilliation')
plt.grid()

ax2 = plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.ylabel('Undamped')
plt.xlabel('time (sec)')

plt.tight_layout()      # 플롯 간의 간격 조정
plt.grid()
plt.show()
```

```python
ax1 = plt.subplot(211)
plt.title('2 Subplot')
ax1.plot(x1, y1, 'yo-')
ax1.set_ylabel('Damped oscilliation')

ax2 = plt.subplot(212)
ax2.plot(x2, y2, 'r.-')
ax2.set_ylabel('Undamped')

plt.xlabel('time (sec)')
ax1.grid(); ax2.grid()
plt.show()
```

```python
np.random.seed(0)

plt.subplot(221)
plt.plot(np.random.rand(5))
plt.title("axes 1")

plt.subplot(222)
plt.plot(np.random.rand(5))
plt.title("axes 2")

plt.subplot(223)
plt.plot(np.random.rand(5))
plt.title("axes 3")

plt.subplot(224)
plt.plot(np.random.rand(5))
plt.title("axes 4")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2)

np.random.seed(0)
axes[0, 0].plot(np.random.rand(5))
axes[0, 0].set_title("axes 1")
axes[0, 1].plot(np.random.rand(5))
axes[0, 1].set_title("axes 2")
axes[1, 0].plot(np.random.rand(5))
axes[1, 0].set_title("axes 3")
axes[1, 1].plot(np.random.rand(5))
axes[1, 1].set_title("axes 4")

plt.tight_layout()
plt.show()
```

```python
# 반복문 이용
np.random.seed(0)

fig, axes = plt.subplots(2, 2) 
for i in range(4):              # i = 0, 1, 2, 3
    ax = axes[i//2, i%2]        # 행 인덱스- i//2, 열 인덱스-i%2
    ax.plot(np.random.rand(5))  # y값 난수 5개/ xtick 임의로 주어짐 0 1 2 3 4
    ax.set_title(f'axes {i+1}')
    ax.grid()
    
plt.tight_layout()
plt.show()
```
