# Folium 지도

```python
import pandas as pd
import folium
```

```python
# OpenStreetMap
folium.Map(location=[37.559868, 126.967109], zoom_start=15)
```

```python
# Stamen Toner Map
folium.Map(location=[37.559868, 126.967109], tiles='Stamen Toner', zoom_start=15)
```

```python
map = folium.Map(location=[37.559868, 126.967109], zoom_start=14)
folium.Marker(
    location=[37.559868, 126.967109],
    popup='한국경제신문사'
).add_to(map)
folium.Marker(
    location=[37.566317, 126.977829],
    tooltip='서울특별시청',
    icon=folium.Icon(color="red", icon="info-sign")
).add_to(map)
folium.Circle(
    location=[37.56412, 126.998009],
    radius=100,
    tooltip='중구청',
    color='crimson',
    fill=False
).add_to(map)
map
```

- addr2.csv를 읽어서 지도위에 표시하기

```python
df = pd.read_csv('addr2.csv')
df
```

```python
# 건물명 tooltip, 주소 popup
map = folium.Map(
    location=[df['위도'].mean(), df['경도'].mean()], 
    zoom_start=13
)
for i in df.index:
    folium.Marker(
        location=[df['위도'][i], df['경도'][i]],
        tooltip=df['건물명'][i],
        popup=folium.Popup(
            df['주소'][i], min_width=200, max_width=200)
    ).add_to(map)
title_html = '<h3 align="center" style="font-size:20px"><b>5개 지역 표시</b></h3>'   
map.get_root().html.add_child(folium.Element(title_html))
map
```
